```python
from collections import defaultdict

import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'Times New Roman'

colors = ['#000000', '#E69F00', '#56B4E9', '#009E73', '#F0E442', '#0072B2', '#D55E00', '#CC79A7']
```

```python
models = ['lxmert', 'vilbert', 'visualbert', 'vl-bert', 'uniter', 'muniter', 'xuniter']
labels = ['LXMERT', 'ViLBERT', 'VisualBERT', 'VL-BERT', 'UNITER', 'mUNITER', 'xUNITER']

langs = ['zh', 'ta', 'sw', 'id', 'tr']
```

## NLVR2

```python
model2en = dict()
for m in models:
    lines = open(f"../experiments/ctrl_{m}/nlvr2/test.err").readlines()
    for l in lines:
        if "Validation" in l:
            val = float(l.split()[-1].strip())
    model2en[m] = val
model2en
```

```python
f, ax = plt.subplots(figsize=(12,8))

mycolors = [colors[1], colors[2], colors[7], colors[6], colors[5], colors[4], colors[3]]
for i, m in enumerate(model2en):
    ax.bar(labels[i], model2en[m], edgecolor='k', color=mycolors[i])

ax.set_ylim(50, 75)
ax.grid(alpha=0.2, axis='y')
ax.set_title('NLVR2', fontsize=32)
ax.set_ylabel('Accuracy', fontsize=28)
ax.tick_params(axis='both', which='major', labelsize=16)

f.savefig('nlvr2.pdf', bbox_inches='tight')
```

## MaRVL

### Zero-Shot Cross-Lingual Transfer

```python
m_model2langs = defaultdict(dict)
for m in models[-2:]:
    for lang in langs:
        lines = open(f"../experiments/ctrl_{m}/marvl/test.{lang}.err").readlines()
        for l in lines:
            if "Validation" in l:
                val = float(l.split()[-1].strip())
            m_model2langs[m][lang] = val
m_model2langs
```

```python
f, ax = plt.subplots(figsize=(12,8))
w = 0.4

ax.bar([i-w/2 for i in range(len(langs))], m_model2langs['muniter'].values(), width=w, color=mycolors[-2], edgecolor='k')
ax.bar([i+w/2 for i in range(len(langs))], m_model2langs['xuniter'].values(), width=w, color=mycolors[-1], edgecolor='k')

ax.axhline(model2en['muniter'], color=mycolors[-2], label='mUNITER')
ax.axhline(model2en['xuniter'], color=mycolors[-1], label='xUNITER')

ax.set_ylim(50, 75)
ax.grid(alpha=0.2, axis='y')
# ax.set_ylabel('Accuracy', fontsize=28)
ax.set_title('MaRVL zero-shot', fontsize=32)
ax.set_xticklabels([''] + langs)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.legend(loc='upper right', fontsize=18, bbox_to_anchor=(0.5, 0., 0.5, .9))

f.savefig('marvl_zs.jpg', bbox_inches='tight')
```

### Translate-Test

```python
en_model2langs = defaultdict(dict)
for m in models[:-2]:
    for lang in langs:
        lines = open(f"../experiments/ctrl_{m}/marvl_translate-test/test.{lang}.err").readlines()
        for l in lines:
            if "Validation" in l:
                val = float(l.split()[-1].strip())
            en_model2langs[m][lang] = val
en_model2langs
```

```python
f, ax = plt.subplots(figsize=(16,8))
w = 0.15

ax.bar([i-2*w for i in range(len(langs))], en_model2langs['lxmert'].values(), width=w, color=mycolors[0], edgecolor='k')
ax.bar([i-w for i in range(len(langs))], en_model2langs['vilbert'].values(), width=w, color=mycolors[1], edgecolor='k')
ax.bar([i for i in range(len(langs))], en_model2langs['visualbert'].values(), width=w, color=mycolors[2], edgecolor='k')
ax.bar([i+w for i in range(len(langs))], en_model2langs['vl-bert'].values(), width=w, color=mycolors[3], edgecolor='k')
ax.bar([i+2*w for i in range(len(langs))], en_model2langs['uniter'].values(), width=w, color=mycolors[4], edgecolor='k')

ax.axhline(model2en['lxmert'], color=mycolors[0], label='LXMERT')
ax.axhline(model2en['vilbert'], color=mycolors[1], label='ViLBERT')
ax.axhline(model2en['visualbert'], color=mycolors[2], label='VisualBERT')
ax.axhline(model2en['vl-bert'], color=mycolors[3], label='VL-BERT')
ax.axhline(model2en['uniter'], color=mycolors[4], label='UNITER')

ax.set_ylim(50, 75)
ax.grid(alpha=0.2, axis='y')
ax.set_ylabel('Accuracy', fontsize=28)
ax.set_title('MaRVL translate-test', fontsize=32)
ax.set_xticklabels([''] + langs)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.legend(loc='upper right', ncol=3, fontsize=18, bbox_to_anchor=(0.05, 0., 0.5, .8))

f.savefig('marvl_translate-test.pdf', bbox_inches='tight')
```

### Performance by category

```python
import pandas as pd
import seaborn as sns
```

```python
colors = ['#0072B2', '#E69F00']

cat2name = {3: "Animal", 5: "Food and Beverages", 6: "Clothing and grooming", 7: "The house", 
            8: "Agriculture and vegetation", 9: "Basic actions and technology", 10: "Motion", 
            14: "Time", 17: "Cognition", 18: "Speech and language", 19: "Religion and belief"}
categories = list(cat2name.keys())
```

```python
def get_vals(d, lang):
    accs = []
    cons = []
    for cat in categories:
        fn = d + f'test.{lang}.cat-{cat}.log'
        with open(fn) as f:
            lines = f.readlines()
            acc = 0
            con = 0
            for l in lines:
                if 'Validation' in l:
                    acc = float(l.split()[-1])
                elif 'Consistency' in l:
                    con = float(l.split()[-1])
            accs.append(acc)
            cons.append(con)
    return accs, cons

def get_val(fn):
    with open(fn) as f:
        lines = f.readlines()
        for l in lines:
            if 'Validation' in l:
                acc = float(l.split()[-1])
            elif 'Consistency' in l:
                con = float(l.split()[-1])
    return acc, con
```

```python
langs = ['id', 'sw', 'ta', 'tr', 'zh']
for l in langs:
    m_accs, m_cons = get_vals(f"../experiments/ctrl_muniter/marvl_by-chapters/", l)
    x_accs, x_cons = get_vals(f"../experiments/ctrl_xuniter/marvl_by-chapters/", l)
```

```python
m_df = pd.DataFrame()
x_df = pd.DataFrame()
for i in [0,1,2,3,4]:
    lang = langs[i]
    m_acc, m_con = get_val(f"../experiments/ctrl_muniter/marvl/test.{lang}.err")
    x_acc, x_con = get_val(f"../experiments/ctrl_xuniter/marvl/test.{lang}.err")
    m_df = m_df.append({"Model": 'mUNITER', "Language": lang, "Chapter": 'ALL', "Accuracy": m_acc, "Consistency": m_con}, ignore_index=True)
    x_df = x_df.append({"Model": 'xUNITER', "Language": lang, "Chapter": 'ALL', "Accuracy": x_acc, "Consistency": x_con}, ignore_index=True)
    m_accs, m_cons = get_vals(f"../experiments/ctrl_muniter/marvl_by-chapters/", lang)
    x_accs, x_cons = get_vals(f"../experiments/ctrl_xuniter/marvl_by-chapters/", lang)
    for ix, cat in enumerate(categories):
        m_df = m_df.append({"Model": 'mUNITER', "Language": lang, "Chapter": cat, "Accuracy": m_accs[ix] if m_accs[ix] > 0 else np.nan, "Consistency": m_cons[ix]}, ignore_index=True)
        x_df = x_df.append({"Model": 'xUNITER', "Language": lang, "Chapter": cat, "Accuracy": x_accs[ix] if x_accs[ix] > 0 else np.nan, "Consistency": x_cons[ix]}, ignore_index=True)

m_df.head()
```

```python
f, axs = plt.subplots(1,2,figsize=(16, 8))
plt.subplots_adjust(wspace=0.1)

ax = axs[0]
df = m_df.pivot('Chapter', 'Language', 'Accuracy')
sns.heatmap(df, annot=True, linewidths=.5, ax=ax, fmt='.1f', cmap='Blues', annot_kws={"fontsize":24}, cbar=False,vmin=0, vmax=80)
ax.set_xticklabels(langs, fontsize=28, rotation=0)
ax.set_yticklabels([cat2name[c] + "  " for c in categories] + ["ALL  "], fontsize=28, rotation=0)
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_title('mUNITER', fontsize=32)

ax = axs[1]
df = x_df.pivot('Chapter', 'Language', 'Accuracy')
sns.heatmap(df, annot=True, linewidths=.5, ax=ax, fmt='.1f', cmap='Blues', annot_kws={"fontsize":24}, cbar=False, vmin=0, vmax=80)
ax.set_xticklabels(langs, fontsize=28, rotation=0)
ax.set_yticklabels("", fontsize=28, rotation=0)
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_yticks([])
ax.set_title('xUNITER', fontsize=32)

f.savefig('marvl_by-chapter.pdf', bbox_inches='tight')
```
